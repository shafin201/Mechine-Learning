**One Hot Encoding**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("Data/loan.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
df["Gender"].fillna(df["Gender"].mode()[0],inplace=True)
```

```python
df["Married"].fillna(df["Married"].mode()[0],inplace=True)
```

**GET Dummy For Encoding**

```python
en_data= df[['Gender','Married']]
en_data
```

```python
pd.get_dummies(en_data)
```

**With ScikitLearn Encoding**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe= OneHotEncoder()
ar= ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=['Gender_Female','Gender_Male','Married_No','Married_Yes'])
```

```python
ohe= OneHotEncoder(drop="first")
ar= ohe.fit_transform(en_data).toarray()

pd.DataFrame(ar,columns=['Gender_Male','Married_Yes'])
```

**Ordinal_Encoding**

```python
data1=pd.read_csv("Data/loan.csv")
```

```python
data1.head()
```

```python
data1["Property_Area"].unique()
```

```python
en_or=[['Rural','Urban','Semiurban']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=en_or)
data1["Property_Area"]= oe.fit_transform(data1[["Property_Area"]])
```
